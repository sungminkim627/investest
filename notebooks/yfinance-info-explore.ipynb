# yfinance .info exploration
Sample tickers: VOO, VFV (ETFs), AAPL, MSFT (stocks).

This notebook fetches `.info` and prints selected fields so we can see what Yahoo returns for each type.

```python
import yfinance as yf
# from pprint import pprint
```

```python
tickers = ["VOO", "VFV.TO", "AAPL", "MSFT"]
infos = {}
for t in tickers:
    print(f"=== {t} ===")
    info = yf.Ticker(t).info
    infos[t] = info
    print(f"keys: {len(info)}")
```
